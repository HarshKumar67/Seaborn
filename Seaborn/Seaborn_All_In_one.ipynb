#### 1. Line plot : with different parameters : sns.lineplot() :

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=pd.read_csv("penguins.csv")
var
```

```python
plt.figure(figsize=(5,5))
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=var ,hue="sex" ,style="sex",
             markers=["o",">"],palette='mako');

## to add a title use: plt.title(""):

plt.title("Seaborn Line plot Graph");

## to add a grid : (i) use plt.grid(), (ii) use sns.set(style="") :

## plt.grid() : a simple grid is bieng created 
sns.set(style="darkgrid") ##: variety of option to create grid
```

### Seaborn Barplot : sns.barplot("-") 

```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd

x=sns.load_dataset("penguins")
```

```python
x
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(5,5))
sns.barplot(x="island", y="bill_length_mm", data=x, hue="sex",errwidth=5,
            palette="Spectral", saturation=100,alpha=1);
```

### Seaborn Scatter plot : sns.scatter("-") :

```python
x
```

```python
plt.figure(figsize=(5,5))
order=["Female","Male"]
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=x, hue="sex", style="sex",
                palette='dark',style_order=order);

## to set x limit : plt.xlim([--]);
plt.xlim([30,60]);
```

### 6. Seaborn Heatmap : made of only matrix :

```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
import numpy as np
```

```python
np.random.seed(43)
plt.figure(figsize=(10,10))
data=np.random.random(size=(4,5));
sns.heatmap(data, annot=True);
```

```python
data
```

```python
## example 2> 
vr=np.linspace(1,10,20).reshape(4,5)
vr
```

```python
## different parameters: 
## vmin, vmax, cmap, annot=True: to make number visible , fmt="s": for string 
plt.figure(figsize=(4,4)) 
sns.heatmap(vr,cmap="summer",annot=True);
plt.title("Seaborn Heat Map");
```

### Seaborn Pair Plot : 

```python
harsh=pd.read_csv("car-sales.csv")
harsh["Price"]=harsh["Price"].str.replace('[/$/,/.]', '', regex=True)
harsh
```

```python
harsh["Price"]=harsh["Price"].astype(int)
```

```python
harsh.dtypes
```

```python
## Note : to control the number of variable use --- vars=["Price", "Odometer (KM)"] --

sns.pairplot(harsh, hue="Make",palette='viridis' , kind="scatter");
```

```python
sns.pairplot(harsh, hue="Make",palette='viridis' , kind="kde", diag_kind="hist");
```

### Seaborn Boxplot :

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
jeje=pd.read_csv("tips.csv")
jeje
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(4,4))
sns.boxplot(x="day", y="total_bill", data=jeje, hue="sex",palette="dark", showmeans=True);
```

```python
plt.figure(figsize=(3,3))
sns.set(style="whitegrid")
sns.boxplot(y=jeje["total_bill"], showmeans=True, color="grey");
```
